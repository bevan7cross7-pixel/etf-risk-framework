# ETF Data Cleaning

This notebook checks and prepares the ETF price data for risk calculations. It examines missing values, date coverage and data quality across the selected ETFs.

```python
import pandas as pd

etf_prices = pd.read_csv(
    "../data/raw/etf_prices_5y.csv",
    parse_dates=["Date"],
    index_col="Date"
)

etf_universe = pd.read_csv(
    "../data/processed/etf_universe_clean.csv"
)

print("Price data shape:", etf_prices.shape)
print("ETF universe shape:", etf_universe.shape)

etf_prices.head()
```

```python
missing_summary = pd.DataFrame({
    "Missing_Values": etf_prices.isna().sum(),
    "Missing_Percentage": (etf_prices.isna().mean() * 100).round(2),
    "First_Available_Date": etf_prices.apply(lambda x: x.first_valid_index()),
    "Last_Available_Date": etf_prices.apply(lambda x: x.last_valid_index())
})

missing_summary.sort_values(
    "Missing_Percentage",
    ascending=False
)
```

```python
clean_prices = etf_prices.ffill()

print("Missing values before:", etf_prices.isna().sum().sum())
print("Missing values after:", clean_prices.isna().sum().sum())

clean_prices.head()
```

```python
clean_prices.index.name = "Date"

clean_prices.to_csv(
    "../data/processed/etf_prices_clean.csv"
)

print("Clean price data saved.")
print("Rows:", clean_prices.shape[0])
print("ETFs:", clean_prices.shape[1])
```
